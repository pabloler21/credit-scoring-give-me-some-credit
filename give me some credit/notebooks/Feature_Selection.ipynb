```python
#se importan las librerias que se van a utilizar
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\pablo\Desktop\proyecto_scoring_gmsc\give me some credit\data\clean\gmsc_clean.csv')
```

```python
corr = df.corr()
```

```python
sns.set_theme(style="white", font_scale=1.05)

plt.figure(figsize=(13,10))

sns.heatmap(
    corr,
    cmap="RdBu_r",
    center=0,
    annot=True,
    fmt=".2f",
    linewidths=.4,
    linecolor="gray",
    cbar_kws={"shrink": .8},
    square=True
)

plt.title("Correlation Matrix  Credit Risk Feature Relationships", fontsize=18, fontweight="bold", pad=15)
plt.tight_layout()
plt.show()
```

```python
for col in df.columns:
    plt.figure(figsize=(5,3))
    sns.kdeplot(df[col], fill=True)
    plt.title(col)
    plt.show()
```
